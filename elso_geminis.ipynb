```python
import pandas as pd
import seaborn as sns

df = sns.load_dataset('titanic')

df['age'].head()
```

```python
subset = df[['pclass', 'sex', 'age']]
print("\n--- Multiple Columns (a DataFrame) ---")
print(subset.head())
```

```python
row_0 = df.loc[0]
print("A single row (also a Series)")
print(row_0.head())
```

```python
first_5_rows = df.loc[0:4]
print(first_5_rows)
```

```python
first_5_columns_and_rows = df.loc[0:4,['age','fare']]
print(first_5_columns_and_rows)
```
